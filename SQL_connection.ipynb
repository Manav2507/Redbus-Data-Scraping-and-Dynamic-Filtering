## CREATE A SQLITE3 DATABASE

```python
#sql connection
# import mysql.connector
import sqlite3

# SQL connection
# conn = mysql.connector.connect(
#     host="localhost",
#     user="root",
#     password="",
#     database = "redbus_database"
# )
# my_cursor = conn.cursor()
# my_cursor.execute("CREATE DATABASE IF NOT EXISTS redbus_database")
# print("Database created successfully!")

conn = sqlite3.connect("redbus_database.db")
```

```python
# Create the table
conn.execute('''
    CREATE TABLE IF NOT EXISTS bus_details(
        Route_Name VARCHAR(225) NULL,
        Route_Link VARCHAR(225) NULL,
        Bus_Name VARCHAR(225) NOT NULL,
        Bus_Type VARCHAR(225) NOT NULL,
        Departing_Time VARCHAR(225) NOT NULL,
        Duration VARCHAR(225) NOT NULL,
        Reaching_Time VARCHAR(225) NOT NULL,
        Star_Rating	Float NULL,
        Price Float NULL,
        Seat_Availability  VARCHAR(225) NOT NULL
        
    )
''')

# conn.close()
print("Table Created Successfully")
```

```python
# Insert the queries

insert_query = ''' INSERT INTO bus_details( 
                    Route_Name,	
                    Route_Link,                  
                    Bus_Name,	
                    Bus_Type,
                    Departing_Time,
                    Duration,	
                    Reaching_Time,	
                    Star_Rating,
                    Price,	
                    Seat_Availability
                    ) 
                    VALUES(?,?,?,?,?,?,?,?,?,?) '''

data = final_df.values.tolist()

# my_cursor.executemany(insert_query, data)
conn.executemany(insert_query, data)

conn.commit()

print("Values inserted Successfully")
# conn.close()
```
